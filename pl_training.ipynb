```python
class SegModel(pl.LightningModule):
    def __init__(self):
        super(SegModel, self).__init__()
        self.batch_size = 8
        self.learning_rate = 1e-3
#         self.net = torchvision.models.segmentation.fcn_resnet50(pretrained = False, progress = True, num_classes = 19)
        self.net = UNet(num_classes = 2, bilinear = False)
#         self.net = torchvision.models.segmentation.deeplabv3_resnet50(pretrained = False, progress = True, num_classes = 19)
        # self.net = ENet(num_classes = 3)
        # self.transform = transforms.Compose([
        #     transforms.ToTensor(),
        #     transforms.Normalize(mean = [0.35675976, 0.37380189, 0.3764753], std = [0.32064945, 0.32098866, 0.32325324])
        # ])
        self.trainset = ForestCoverDataset(mode='train', one_hot_masks=True)
        self.testset = ForestCoverDataset(mode='val', one_hot_masks=True)
        
    def forward(self, x):
        return self.net(x)
    
    def training_step(self, batch, batch_nb) :
        print(batch_nb)
        img, mask = batch
        print(img.shape)
        print(mask.shape)
        # img = img.float()
        # mask = mask.float()
        # mask = mask.squeeze(1)
        # print(mask.shape)
        out = self.forward(img)
        print(out.shape)
        loss_val = F.cross_entropy(out, mask)
#         print(loss.shape)
        return loss_val
    
    def configure_optimizers(self):
        opt = torch.optim.Adam(self.net.parameters(), lr = self.learning_rate)
        # sch = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max = 10)
        # return [opt], [sch]
        return opt
    
    def train_dataloader(self):
        return DataLoader(self.trainset, batch_size = self.batch_size, shuffle = True, num_workers = 4)
    
    def test_dataloader(self):
        return DataLoader(self.testset, batch_size = 1, shuffle = False, num_workers = 4)
```

```python
model = SegModel()
trainer = pl.Trainer(max_epochs=5, profiler='simple', accelerator='cpu')

# checkpoint_callback = pl.callbacks.ModelCheckpoint(
#     filepath = 'checkpoints_enet/',
#     save_best_only = True,
#     verbose = True, 
#     monitor = 'loss',
#     mode = 'min',
#     prefix = ''
# )
# trainer = pl.Trainer(gpus = 1, max_nb_epochs = 70, checkpoint_callback = checkpoint_callback, early_stop_callback = None)
trainer.fit(model)
```
